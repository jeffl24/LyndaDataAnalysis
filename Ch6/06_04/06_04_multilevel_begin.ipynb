```python
# compound labels miltiindex
# index dataframes with multiindex
import pandas as pd
import seaborn
```

```python
flights = seaborn.load_dataset('flights')
flights.head()
```

```python
flights_indexed = flights.set_index(['year', 'month'])
flights_indexed.head()
# bold typefaces mean that they comprise multi-index
```

```python
flights_indexed.loc[1951].loc["January": "June"]
# loc function will only select rows based on 
# the first levels of the multi-level index
```

```python
# unstack() transposes the second level of index into columns
flights_unstacked = flights_indexed.unstack()
flights_unstacked
```

```python
# perform SUM along the rows
flights_unstacked.sum(axis=1)
```

```python
flights_unstacked['month']
```

```python
flights_restacked = flights_unstacked.stack()
flights_restacked.tail(20)
```

```python
flights_restacked.index
```

```python
flights_restacked[flights_restacked.index.get_level_values('1952').isin([390])]
```
