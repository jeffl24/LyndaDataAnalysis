```python
# remove \n from each dictionary entry
# convert all words into lowercase
wordlist_clean = [dictionary.strip().lower() for dictionary
                                              in wordlist]
wordlist_clean[:10]
```

```python
#to remove duplicates,
# wrap the list into a set
wordlist_clean_set = list(set(wordlist_clean))
wordlist_clean_set[:10]
```
